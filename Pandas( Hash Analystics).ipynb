### Adding New Column


```python
df['Range'] = np.arange(0,(len(df)*2),2)

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Speed'] + df['Sp. Def'] + df['Sp. Atk']
# same as
#df['Total_index'] = df.iloc[:, 4:10].sum(axis=1)

df.head(10)
```

### Accessing Rows and Columns

Selecting specific [rows, columns] using 'loc', 'iloc', 'at', & 'iat'

```python
df.iloc[0:5, 3:10]
# same as
# df.loc[0:5, 'HP':'Speed']

# df.loc[5, 'Type 1':'Type 2']

# df.loc[:, 'Name':'Type 2' ]
# same as
# df.iloc[:, 0:3]


#using loc.iloc, at, iat
df.loc[2, 'Legendary']
# same as
df.iloc[2, 10]
# same as
df.at[2, 'Legendary']
#same as
df.iat[2, 10]
```

### Data Filtering

```python
df[df['Legendary'] == True].head()
```

```python
filtered = df[(df['Type 1'] == 'Ice') | (df['Type 2'] == 'Dragon') & (df['HP'] > 100)]
filtered.head(10)
```

```python

df[~df['Name'].str.contains('Mega')].head()
# ~ means "NOT" 
```

```python
df[df['Name'].str.contains('^Pi')].head()
# symbol ('^') means starts_with
```

```python
import re #regular expressions

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head()
```

```python
# isin

df[df['Type 2'].isin(['Poison', 'Flying'])].head()
```

### Missind Data
1. dropna()
2. fillna()
3. isnull()

```python
# DROPNA

# dropna(how="any"/"all"), it is "any" by default, meaning it will drop any row that has Nan value, 
# "all": will only drop a row if all the row value has Nan Value
df.dropna(how="any")
# OR
df.dropna()

# FILLNA

df.fillna(value=20)

# ISNA
df.isnull()

# NOTNULL
df.notnull().head()
```
